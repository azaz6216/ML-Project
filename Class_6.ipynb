```python
import pandas as pd
df= pd.read_csv('/content/drive/MyDrive/Profit Prediction.csv')
df
```

```python
df.isnull().sum()
```

```python
areadummies = pd.get_dummies(df['Area']).astype(int)
df
```

```python
areadummies.head()
```

```python
df['Area'].value_counts()
```

```python
df2=pd.concat([df,areadummies],axis=1)
```

```python
df2.head(10)
```

```python
x= df2.drop(['Area','Profit'],axis = 1)
y=df2['Profit']

```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
xtrain.head(3)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
```

```python
xtrain_mms = mms.fit_transform(xtrain)
xtest_mms = mms.fit_transform(xtest)
```

```python
print (xtrain_mms)
```

```python
xtrain_mms_df = pd.DataFrame(xtrain_mms,columns = xtrain.columns)
xtest_mms_df = pd.DataFrame(xtest_mms,columns=xtest.columns)
```

```python
from sklearn.linear_model import LinearRegression
model_linear = LinearRegression()
model_linear.fit(xtrain_mms_df,ytrain)
```

```python
model_linear.score(xtest_mms_df,ytest)
```

```python
xtest.head()
```

```python
from sklearn import metrics
import numpy as np
test_pred = model_linear.predict(xtest_mms_df)
mae=metrics.mean_absolute_error(ytest,test_pred)
mse=metrics.mean_squared_error(ytest,test_pred)
print('MAE:' ,mae)
print('MSE:' ,mse)
```
