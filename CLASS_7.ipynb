```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
df
```

```python
X=df
```

```python
plt.scatter(x = 'Annual Income (k$)' , y = 'Spending Score (1-100)' , data = df)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,20):
  Kmeans=KMeans(n_clusters=i)
  Kmeans.fit(X)
  wcss.append(Kmeans.inertia_)
plt.figure(1,figsize = (10,6))
plt.plot(np.arange(1 , 20) , wcss , 'o')
plt.plot(np.arange(1 , 20) , wcss ,'-' , alpha = 0.5)
plt.xlabel('Number of cluster') , plt.ylabel('Wcss/Inertia')
plt.show()
```

```python
Kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42) # Changed 'K-means++' to 'k-means++'
Kmeans = Kmeans.fit(X)
ridi = Kmeans.fit_predict(X)
print(ridi)
```

```python
test=Kmeans.predict(np.asarray([[3,3]]))
test[0]
```

```python
inertia_score = Kmeans.inertia_
inertia_score
```

```python
from sklearn.metrics import silhouette_score
Kmeans_labels = ridi
silhouette_avg = silhouette_score(X, Kmeans_labels)
silhouette_avg
```

```python
algorithm=Kmeans
labels2 = algorithm.labels_
centroids2 = algorithm.cluster_centers_


```

```python
print("Cluster Labels:")
print(labels2)
print("\nCentroids:")
print(centroids2)
```

```python
X2=np.array(X)
h = 0.02
x_min, x_max = X2[:, 0].min() - 1, X2[:, 0].max() + 1
y_min, y_max = X2[:, 1].min() - 1, X2[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z2 = algorithm.predict(np.c_[xx.ravel(), yy.ravel()])


plt.figure(1 , figsize = (10 , 7) )
plt.clf()
Z2 = Z2.reshape(xx.shape)
plt.imshow(Z2 , interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap = plt.cm.Pastel2, aspect = 'auto', origin='lower')

plt.scatter( x = 'Annual Income (k$)' ,y = 'Spending Score (1-100)' , data = df , c = labels2 ,
            s = 200 )
plt.scatter(x = centroids2[: , 0] , y =  centroids2[: , 1] , s = 300 , c = 'red' , alpha = 0.5)
plt.ylabel('Spending Score (1-100)') , plt.xlabel('Annual Income (k$)')
plt.show()
```
