```python
# Import pandas as pd
import pandas as pd
```

```python
# read in heart.csv
df=pd.read_csv(r"https://raw.githubusercontent.com/azaz6216/dataset/refs/heads/main/heart.csv")
```

```python
# Display the head of the dataframe
df.head()
```

```python
# Print the info 
df.info()
```

```python
# Print a summary of the dataset
df.describe()
```

```python
# Display the columns of the dataframe
df.columns
```

```python
# Create the X and y variables
X=df.drop(['target'],axis=1)
y=df['target']
```

```python
# Split the dataset into a training and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# Import the DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create an instance of the DecisionTreeClassifier
tree=DecisionTreeClassifier()
```

```python
# Fit the DecisionTreeClassifier to the training set
tree.fit(X_train,y_train)
```

```python
# Make predictions using the DecisionTreeClassifier
predictions=tree.predict(X_test)
```

```python
# Import the classification metrics
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
# Print the accuracy of the DecisionTreeClassifier
print(accuracy_score(predictions,y_test))
```

```python
# Display the confussion matrix
print(confusion_matrix(predictions,y_test))
```

```python
# Plot the confussion matrix
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(confusion_matrix(predictions,y_test),annot=True)
```

```python
# Print the classification report 
print(classification_report(predictions,y_test))
```

```python
# Import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create an instance of the RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
# Make predictions using the RandomForestClassifier
predictions=rf.predict(X_test)
```

```python
# Print the accuracy of the RandomForestClassifier
print(accuracy_score(predictions,y_test))
```

```python
# Print the confusion matrix
q=confusion_matrix(predictions,y_test)
q
```

```python
# Plot the confusion the confusion matrix
plt.figure(figsize=(12,6))
sns.heatmap(q,annot=True)
```

```python
# Print the classification report
print(classification_report(predictions,y_test))
```
